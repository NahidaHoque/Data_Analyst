# Replace and change data types

```python
# mixed data thakle data replace korte hbe
# replace korle onk time e data type change korte hbe
# mixed data = dependents column e - 0, 1, 2 , 3+ ache.. 3+ holo categorical data
# so 3+ k replace korbo numerical data te then convert korbo int data type
```

```python
df4 = pd.read_csv('loan.csv')
```

```python
#data type check korbo
df4.info()
```

```python
# dependents e null value ache and categorical value ache
df4.isnull().sum()
```

```python
# kon data kotobr ache
df4["Dependents"].value_counts()
```

```python
# missing data mode diye fill korbo, eta discrete typer data numerical na so mode diye fill korbo
df4["Dependents"].fillna(df4["Dependents"].mode()[0],inplace = True)
```

```python
df4.isnull().sum()
```

```python
# replace korte hbe +3 k 3
df4["Dependents"].replace("3+","3",inplace = True)
```

```python
df4["Dependents"].value_counts()
```

```python
df4.info()
```

```python
# data type change korbo
df4["Dependents"] = df4["Dependents"].astype("int64")
```

```python
df4.info()
```
