# One hot encoding and dummy variable (Ordinal Encoding)

```python
# encoding : categorical data k numerical data te convert korte chai
```

```python
#One hot encoding: limited data r upor kora hy. Like Gender: Male , Female ; Married: Yes, No ; Property: urban, semiurban, rural
# Gender: Male , Female ---> Gender_male, Gender_female emn 2ta new col hy and male hole 0 nahole 1 boshe
# evabe categorical data numerical data te convert hy
```

```python
import pandas as pd
```

```python
df = pd.read_csv('loan.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df["Gender"].fillna(df["Gender"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Married"].fillna(df["Married"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
#One hot encoding: get dummies() - pandas er method
```

```python
#Gender r Married k alada korbo
en_data = df[["Gender","Married"]]
```

```python
en_data
```

```python
#encoding
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
#scikitlearn library use kore one hot coding
#scikit-learn : preprocessing, modeling
from sklearn.preprocessing import OneHotEncoder
```

```python
#fit_transform() : transform kore mane categorical data k numerical data te convert kore
ohe = OneHotEncoder(drop = "first")
ohe.fit_transform(en_data)
```

```python
#sparse matrix: jekhane beshirvag content 0 or 1 diye fill hy. so data encode hye jay
```

```python
# data ta k ekta array format e anbo
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
#array ta k dataframe e convert korte hbe
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```
