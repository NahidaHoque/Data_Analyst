# Ordinal Encoding

```python
# ordinal encoding 2 vabe kora jay - 1. sklearn  2. Map function
```

```python
df2 = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
```

```python
df2
```

```python
df2.head(3)
```

```python
#sequestial data toiri korlam
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df2[["Size"]])
```

```python
oe.transform(df2[["Size"]])
```

```python
df2["Size_en"] = oe.transform(df2[["Size"]])
```

```python
df2
```

```python
# map function use kore ordinal encoding
df3 = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
```

```python
# ekhane iccchamoto number dewa jay
ord_data1 = {"s":1,"m":2,"l":3,"xl":4}
```

```python
df3["Size"].map(ord_data1)
```

```python
df3["Size_en_map"] = df3["Size"].map(ord_data1)
```

```python
df3
```

```python
# map function : icchamoto number dewa jay, machine learning algo kono kisur output improve er jnno er number change kore kichuta ta improve kora jay
```
